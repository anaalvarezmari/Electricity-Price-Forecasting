```python
#def function split time stamp
def  split_time_stamp(table):
## split time stamp
    table['Dates'] = pd.to_datetime(table['settlementdate'],dayfirst=True).dt.date
    table['Time'] = pd.to_datetime(table['settlementdate'],dayfirst=True).dt.time
    table['year'] = pd.DatetimeIndex(table['Dates'],dayfirst=True).year
    table['month'] = pd.DatetimeIndex(table['Dates'],dayfirst=True).month
    table['day_of_week'] = pd.DatetimeIndex(table['Dates'],dayfirst=True).dayofweek
    return table
```

```python
demandData=merged_data
demandData=split_time_stamp(demandData)
```

```python
def time_to_num(time_str):
    hh,mm,ss=map(int,time_str.split(':'))
    return (mm+60*hh)/5
```
