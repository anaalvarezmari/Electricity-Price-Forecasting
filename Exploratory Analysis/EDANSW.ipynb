```python
#!/usr/bin/python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset=pd.read_csv(r'C:\Users\anaal\datafinal.csv')
dataset['timeIndex']= (dataset['newTime']+1)/288
```

```python
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 9}

matplotlib.rc('font', **font)
fig, ax = plt.subplots(figsize=(10, 6))
plt.scatter(dataset['rrp_y'],dataset['rrp_x'],cmap='viridis')
plt.yticks(np.arange(-1000, 15000, 1000))
plt.xticks(np.arange(-1000, 15000, 1000))

plt.title("Price QLD vs Price NSW") 
plt.xlabel("Price NSW (AUD)")
plt.ylabel("Price QLD (AUD)")
plt.show()
```
